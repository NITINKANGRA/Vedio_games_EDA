#**Exploratory Data Analysis  & Data Visualization**

# **1.What are the different Genre's of vedio games ?**

```python
value_counts_genre = df['Genre'].value_counts()
value_counts_genre
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=value_counts_genre.index, y=value_counts_genre.values, palette='viridis')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **2.What are the different measures of frequency distribution of year trends ?**

```python
frequency_distribution_year = df['Year'].value_counts()
frequency_distribution_year
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(13, 6))
sns.barplot(x=frequency_distribution_year.index, y=frequency_distribution_year.values)
plt.title('Frequency Distribution of Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

# **3.What are the top 10 publishers of vedio games industry ?**

```python
value_counts_publisher = df['Publisher'].value_counts()
top_10_publishers = value_counts_publisher.head(10)
top_10_publishers
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top_10_publishers.values, y=top_10_publishers.index, palette='hls')
ax.set(xlabel='Count', ylabel='Publisher')
ax.set_title('Top 10 Publishers')
for i, v in enumerate(top_10_publishers.values):
    ax.text(v + 3, i, str(v), color='black', fontweight='bold')
plt.tight_layout()
plt.show()
```

# **4.What is the Global sales of top 10 publishers ?** 

```python
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum()
publisher_sales
```

```python
top_10_publishers = publisher_sales.nlargest(10)
top_10_publishers
```

```python
plt.figure(figsize=(10, 6))
plt.pie(top_10_publishers, labels=top_10_publishers.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Top 10 Publishers by Global Sales')
plt.show()
```
